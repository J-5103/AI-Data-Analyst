```python
from langgraph.graph import StateGraph , START , END
from langchain_groq import ChatGroq
from langchain_experimental.agents import create_pandas_dataframe_agent
from langchain_core.tools import tool
from typing import TypedDict
import pandas as pd 
import matplotlib.pyplot as plt
import matplotlib
matplotlib.use('Agg')
import os
import warnings
warnings.filterwarnings('ignore')

os.makedirs("outputs" , exist_ok= True)
```

LLM AND API


```python

os.environ["GROQ_API_KEY"] = "YOUR_GROQ_API"

llm = ChatGroq(
    model="llama-3.1-8b-instant",
    temperature= 0
)
```

CREATE SAMPLE SALES DATA

```python

import pandas as pd
import numpy as np

np.random.seed(42)

data = {
    "Month": ["jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"] * 3,
    "Product": (["Laptop"]*12) + (["Mobile"]*12) + (["Tablet"]*12),
    "Sales": np.random.randint(50,300,36),
    "Revenue": np.random.randint(50000,500000,36),
    "Region": np.random.choice(["North", "South" , "East","West"] , 36),
    "Units_sold": np.random.randint(10,200,36)
}

df = pd.DataFrame(data)
df.to_csv("data/sales_data.csv",index = False)
print("✅ Sample data created!")
print(df.head())
```

DEFINE LANGGRAPH STATE

```python
class AnalystState(TypedDict):
    file_path:str
    df_summary : str
    analysis_result:str
    chart_path:str
    final_report:str
    user_question:str
```

NODE 1 : LOAD AND SUMMARIZE DATA

```python
def load_data_node(state:AnalystState) -> AnalystState:
    print("📂 Node 1: Loading data...")

    df = pd.read_csv(state["file_path"])

    summary = f"""
    Dataset Loaded Successfully!
    Shape : {df.shape[0]} rows x {df.shape[1]} colmmns
    Columns : {list(df.columns)}
    Data Types: {df.dtypes.to_dict()}
    Missing : {df.isnull().sum().to_dict()}
    sample :
    {df.head(3).to_string()}
    """

    print("✅ Data loaded!")
    return {"df_summary": summary}
```

NODE 2 : ANALYZE WITH PANDAS AGENT

```python
def analyze_data_node(state:AnalystState) -> AnalystState:
    print("🤖 Node 2: Analyzing data with AI...")

    df = pd.read_csv(state["file_path"])

    agent = create_pandas_dataframe_agent(
        llm,
        df,
        agent_type="tool-calling",
        verbose=True,
        allow_dangerous_code=True
    )

    question = state.get("user_question") or """
        perform a complete analysis:
        1. which product has the highest total revenue?
        2. which region perform best in sales?
        3. what is the monthly sales trend?
        4. top 3 insights from this data?
    """

    result = agent.invoke({"input":question})
    analysis = result.get("output" , str(result))

    print("✅ Analysis done!")
    return {"analysis_result":analysis}
```

NODE 3 : GENERATE CHARTS

```python
def visualize_data_node(state:AnalystState)-> AnalystState:
    print("📊 Node 3: Creating charts...")

    df = pd.read_csv(state["file_path"])

    fig = plt.figure(figsize=(14, 10))
    fig.suptitle("📊 AI Data Analyst — Sales Dashboard", fontsize=16, fontweight='bold')
    
    ax1 = fig.add_subplot(2, 2, 1)
    ax2 = fig.add_subplot(2, 2, 2)
    ax3 = fig.add_subplot(2, 2, 3)
    ax4 = fig.add_subplot(2, 2, 4)


    #chart 1 --> revenue by product

    product_rev = df.groupby("Product")["Revenue"].sum()
    ax1.bar(product_rev.index , product_rev.values,
                  color = ["#4CAF50", "#2196F3", "#FF9800"])
    ax1.set_title("Total Revenue by Product")
    ax1.set_ylabel("Revenue (₹)")

    #chart 2 --> sales by region(pie)

    region_sales = df.groupby("Region")["Sales"].sum()
    ax2.pie(region_sales.values , labels = region_sales.index,
                  autopct = "%1.1f%%" , startangle = 90)
    ax2.set_title("Sales Distribution By Region")

    #chart - 3 --> monthly revenue trend

    month_order =  ["Jan","Feb","Mar","Apr","May","Jun",
                   "Jul","Aug","Sep","Oct","Nov","Dec"]
    monthly = df.groupby("Month")["Revenue"].sum().reindex(month_order)
    ax3.plot(monthly.index , monthly.values,
                   marker = "o" , color="#E91E63", linewidth=2)
    ax3.set_title("Monthly Revenue Trend")
    ax3.set_ylabel("Revenue (₹)")
    ax3.tick_params(axis='x',rotation = 45)


    #chart - 4 --> units sold by product

    units = df.groupby("Product")["Units_sold"].sum()
    ax4.barh(units.index , units.values,
                   color = ["#9C27B0", "#00BCD4", "#FF5722"])
    ax4.set_title("Total Units Sold by Product")
    ax4.set_xlabel("Units")

    plt.tight_layout()
    
    chart_path = "outputs/sales_dashboard.png"
    plt.savefig(chart_path , dpi = 150 , bbox_inches = "tight")
    plt.close()

    print(f"✅ Charts saved to {chart_path}!")
    return {"chart_path": chart_path}
                 

```

NODE 4: GENERATE FINAL REPORT

```python
def generate_report_node(state:AnalystState) -> AnalystState:
    print("📝 Node 4: Generating final report...")

    prompt = f"""
    You are a professional data analyst. Based on the following information,
    write a clean, structured business report.

    DATA SUMMARY:
    {state["df_summary"]}

    AI ANALYSIS RESULTS:
    {state["analysis_result"]}

    write a report with the sections:
    1. Executive Summary
    2. Key Findings
    3. Product Performance
    4. Regional Performance
    5. Recommendations

    keep it professional and concise.

    """

    report = llm.invoke(prompt).content

    with open("outputs/report.txt" , "w") as f:
        f.write(report)

    print("✅ Report generated and saved!")   
    return {"final_report" : report} 

```

BULDING LANGGRAPH WORKFLOW

```python
def build_analyst_graph():
    graph = StateGraph(AnalystState)

    graph.add_node("load_data" , load_data_node)
    graph.add_node("analyze_data" , analyze_data_node)
    graph.add_node("visualize_data" , visualize_data_node)
    graph.add_node("generate_report" , generate_report_node)

    graph.add_edge(START , "load_data")
    graph.add_edge("load_data" , "analyze_data")
    graph.add_edge("analyze_data" , "visualize_data")
    graph.add_edge("visualize_data" , "generate_report")
    graph.add_edge("generate_report" , END)

    return graph.compile()

workflow = build_analyst_graph()
print("✅ Graph compiled successfully!")

```

RUN PROJECT

```python

initial_state = {
    "file_path"     : "data/sales_data.csv",
    "df_summary"    : "",
    "analysis_result": "",
    "chart_path"    : "",
    "final_report"  : "",
    "user_question" : "Which product generates the most revenue and which region has highest sales?"
}

print("🚀 Starting AI Data Analyst...\n")
result = workflow.invoke(initial_state)

print("\n" + "="*60)
print("📊 DATA SUMMARY")
print("="*60)
print(result["df_summary"])


print("\n" + "="*60)
print("🤖 AI ANALYSIS")
print("="*60)
print(result["analysis_result"])

print("\n" + "="*60)
print("📝 FINAL REPORT")
print("="*60)
print(result["final_report"])

print(f"\n✅ Chart saved at : {result['chart_path']}")
print("✅ Report saved at: outputs/report.txt")
```
